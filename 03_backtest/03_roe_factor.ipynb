```python
import akshare as ak
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

# 读取之前的收盘价数据
close_df = pd.read_csv("../03_backtest/close_data.csv",
                       index_col=0, parse_dates=True)

# 只取深市股票
sz_stocks = [code for code in close_df.columns
             if code.startswith(("000", "002", "300"))]

print(f"深市股票数量：{len(sz_stocks)}")
```

```python
roe_dict = {}

for i, code in enumerate(sz_stocks):
    try:
        df_temp = ak.stock_financial_analysis_indicator(
            symbol=code, start_year="2020")
        roe_temp = df_temp[["日期", "净资产收益率(%)"]].copy()
        roe_temp["日期"] = pd.to_datetime(roe_temp["日期"])
        roe_temp = roe_temp.set_index("日期")
        roe_temp.columns = ["ROE"]
        
        # TTM化
        roe_temp["季度"] = roe_temp.index.month.map(
            {3:1, 6:2, 9:3, 12:4})
        roe_temp["ROE_TTM"] = roe_temp["ROE"] / roe_temp["季度"] * 4
        
        # 延迟45天
        roe_temp.index = roe_temp.index + pd.Timedelta(days=45)
        
        # 填充到交易日
        roe_daily = roe_temp[["ROE_TTM"]].reindex(
            close_df.index, method="ffill")
        roe_dict[code] = roe_daily["ROE_TTM"]
        
        if i % 20 == 0:
            print(f"进度：{i}/{len(sz_stocks)}")
        
        time.sleep(0.3)
    except:
        print(f"获取失败：{code}")
        continue

roe_df = pd.DataFrame(roe_dict)
roe_df = roe_df[roe_df.index >= "2023-01-01"]
roe_df.to_csv("roe_data.csv")
print(f"\n完成，维度：{roe_df.shape}")
```

```python
# 对之前失败的股票加重试
failed = [code for code in sz_stocks if code not in roe_df.columns]
print(f"失败股票数：{len(failed)}")

roe_dict2 = {}
for i, code in enumerate(failed):
    for attempt in range(3):  # 最多重试3次
        try:
            df_temp = ak.stock_financial_analysis_indicator(
                symbol=code, start_year="2020")
            roe_temp = df_temp[["日期", "净资产收益率(%)"]].copy()
            roe_temp["日期"] = pd.to_datetime(roe_temp["日期"])
            roe_temp = roe_temp.set_index("日期")
            roe_temp.columns = ["ROE"]
            roe_temp["季度"] = roe_temp.index.month.map(
                {3:1, 6:2, 9:3, 12:4})
            roe_temp["ROE_TTM"] = roe_temp["ROE"] / roe_temp["季度"] * 4
            roe_temp.index = roe_temp.index + pd.Timedelta(days=45)
            roe_daily = roe_temp[["ROE_TTM"]].reindex(
                close_df.index, method="ffill")
            roe_dict2[code] = roe_daily["ROE_TTM"]
            print(f"重试成功：{code}")
            break
        except:
            time.sleep(1)
            continue

roe_df2 = pd.DataFrame(roe_dict2)
roe_df2 = roe_df2[roe_df2.index >= "2023-01-01"]

# 合并
roe_df_all = pd.concat([roe_df, roe_df2], axis=1)
roe_df_all.to_csv("roe_data.csv")
print(f"\n合并后维度：{roe_df_all.shape}")
```

```python
# 读取ROE数据
roe_df_all = pd.read_csv("roe_data.csv", index_col=0, parse_dates=True)
roe_df_all = roe_df_all[roe_df_all.index >= "2023-01-01"]

# 未来20天涨幅
close_sz = close_df[[c for c in roe_df_all.columns if c in close_df.columns]]
future_return = close_sz.shift(-20) / close_sz - 1

# 每天计算截面IC
ic_roe = []
dates_roe = []

for date in roe_df_all.index:
    factor_row = roe_df_all.loc[date].dropna()
    if date not in future_return.index:
        continue
    return_row = future_return.loc[date].dropna()
    common = factor_row.index.intersection(return_row.index)
    if len(common) < 10:
        continue
    ic = factor_row[common].corr(return_row[common], method="spearman")
    ic_roe.append(ic)
    dates_roe.append(date)

ic_roe = pd.Series(ic_roe, index=dates_roe)

print(f"ROE因子 IC均值：{ic_roe.mean():.4f}")
print(f"ROE因子 IC标准差：{ic_roe.std():.4f}")
print(f"ROE因子 ICIR：{ic_roe.mean()/ic_roe.std():.4f}")
print(f"ROE因子 IC>0的比例：{(ic_roe>0).mean():.2%}")
```
